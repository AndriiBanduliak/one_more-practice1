```python
import pandas as pd
import matplotlib.pyplot as plt

# загрузка данных из файла csv
data = pd.read_csv('laptops_10.csv', encoding='ISO-8859-1')

# распределение типов носителя
plt.hist(data['Memory_Type'], bins=10)
plt.xlabel('Storage Type')
plt.ylabel('Count')
plt.title('Storage Type Distribution')

# распределение компаний производителей
plt.hist(data['Company'], bins=10)
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.title('Manufacturer Distribution')

# распределение операционной системы
plt.hist(data['OpSys'], bins=10)
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.title('Operating System Distribution')

# распределение компаний производителей CPU
plt.hist(data['Cpu'], bins=10)
plt.xlabel('CPU Manufacturer')
plt.ylabel('Count')
plt.title('CPU Manufacturer Distribution')

# отображение графиков
plt.show()
```

3.1 Постройте график в абсолютных величинах

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
data = pd.read_csv('laptops_10.csv')

# Удаление строк, в которых компания-производитель процессора является Samsung
data = data[data['Company'] != 'Samsung']

# Подсчет количества ноутбуков по каждой компании-производителю и каждой компании-производителю процессора
counts = data.groupby(['Company', 'Cpu_Company']).size().unstack(fill_value=0)

# Построение многорядовой столбчатой диаграммы в абсолютных величинах
counts.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Laptop Manufacturer')
plt.ylabel('Count')
plt.title('Relationship between Laptop and CPU Manufacturers')
plt.legend(title='CPU Manufacturers')
plt.show()
```
